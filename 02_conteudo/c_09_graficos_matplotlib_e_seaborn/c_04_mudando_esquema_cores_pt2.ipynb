## Gráfico de barras empilhadas para mostrar a distribuição de diferentes tipos de vendas ao longo do tempo

Considere que você tenha uma empresa e queira mostrar a distribuição de diferentes tipos de vendas ao longo do tempo. Crie um gráfico de barras empilhadas:

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [11, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

# Gráficos de colunas sobrescritos
plt.bar(anos, vendas_online)
plt.bar(anos, vendas_loja)
plt.show()  
```

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

# Gráficos empilhados de colunas
# A utilização do parâmetro label indica o nome que será utilizado na legenda - rótulo
plt.bar(anos, vendas_online, label='Online')
# Para definir o empilhamento, utiliza-se o bottom = ...
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()  
```

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

# forçar os anos a serem exibidos como inteiros
# ticks é o "espaçamento", e utiliza as a variável anos como parâmetro
# xticks indica para o eixo x e yticks para o y.
plt.xticks(ticks=anos, labels=anos)
# Caso se quisesse alterar os nomes dos anos, seria possível passar da seguinte forma:
plt.xticks(ticks=anos, labels=['ano1', 'ano2', 'ano3', 'ano4'])

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
# Exibe a legenda
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

# alterar as cores padrão:
# Nome dentro das aspas deve estar conforme documentação da biblioteca
cores = plt.get_cmap('Pastel2').colors

# É possível pegar cada cor individualmente, utilizando o seguinte formato:
plt.bar(anos, vendas_online, label='Online', color=cores[0])
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja', color=cores[1])

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos, labels=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Para garantir que o matplotlib repita as cores, é necessário importar o cycler
# Isto ocorre para que não seja necessário alterar manualmente cada cor
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

"""
rc = run configuration (configuração quando executa) - quando estiver criando o gráfico,
usa as instruções definidas em rc

Utilizado para configurar a paleta de cores como padrão para todos os gráficos
"""
# Neste caso: Quando estiver criando o sistema de eixo, utiliza o seguinte ciclo de propriedads
plt.rc('axes', prop_cycle=(cycler('color', plt.get_cmap('Pastel2').colors)))
 
plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```
