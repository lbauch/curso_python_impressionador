```python
import matplotlib.pyplot as plt

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [70, 45, 23, 88]
destaque = [0.05, 0, 0, 0]  

def formatar_autopct(pct, quantidades):
    total = sum(quantidades)
    valor_absoluto = int(round(pct * total / 100.0))
    return f'{pct:.1f}%\n({valor_absoluto})'

fig, ax = plt.subplots()

ax.pie(
    quantidades, 
    labels=categorias, 
    explode=destaque, 
    autopct=lambda pct: formatar_autopct(pct, quantidades)
)

ax.set_title('Categorias de produtos vendidos')

plt.show()
```
